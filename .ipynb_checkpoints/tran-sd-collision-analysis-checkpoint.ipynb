```python
#Import dependencies

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import iqr
from scipy.stats import linregress
import numpy as np
```

```python
# Read in the collision violations data set
collision = pd.read_csv('pd_collisions_datasd_v1.csv')
collision.head()
```

#What are the most common collision violations in general and by zip code?
#Bar graph of % of collision violations
#Assigned to: Tran


```python
#drop column with null data
# pandas drop columns using list of column names
collision_filter = collision[['violation_section', 'charge_desc']]
```

```python
#make into a new data frame
collision_df = pd.DataFrame(collision_filter) 
collision_df
```

```python
pie = collision_df.violation_section.value_counts()
pie.head()
```

```python
#create summary bar data frame
pie_df = pd.DataFrame({
    "count": pie,
    
    })
pie_df
```

```python
#sort top 7 collision 
top_seven_collision = pie_df.nlargest(7,'count')
top_seven_collision
```

```python
#pie chart#counter clockwise
colors = ["#E13F29", "#D69A80", "#D63B59", "#AE5552", "#CB5C3B", "#EB8076", "#96624E"]

# Create a pie chart
plt.pie(
    # using data total collision
    top_seven_collision['count'],
    # with no shadows
    shadow=False,
    # with colors
    colors=colors,
    # with the start angle at 200
    startangle=200,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )
    #label
labels = ['Signal Required', 'Misc Hazardous', 
'Speeding', 'Parking or Backing Randomly','Running Stop Sign','Tailing','Left Turn or U-Turn']

# View the plot drop above
plt.axis('equal')
plt.legend(labels,bbox_to_anchor=(0.9,0.3), loc="center", fontsize=9, 
           bbox_transform=plt.gcf().transFigure)

# View the plot
plt.title('Top Seven Collision in San Diego')
plt.tight_layout()
plt.savefig('pie_plot.png',bbox_inches="tight")
plt.show()
```

```python
# Read in the collision violations data set
zipcode_df = pd.read_csv('zipcode.csv')
zipcode_df.head()
```

```python
#split zipco
zipcode_df['zipcode'] = zipcode_df['Full_Address'].str.extract(r'(\d{5}\-?\d{0,4})')
zipcode_df
```

```python
#convert zipcode column into int for coding manipulation
zipcode_df.zipcode = pd.to_numeric(zipcode_df.zipcode, errors='coerce').fillna(0).astype(np.int64)
print(zipcode_df)
```

```python
#sort out CA zip
ca_zip = zipcode_df[zipcode_df.zipcode > 90000]
ca_zip
```

```python
#merge zipcode column
ca_zip_df = collision_df.assign(zipcode=ca_zip['zipcode'])
```

```python
#drop na values
ca_zip_df = ca_zip_df[ca_zip_df['zipcode'].notna()]
ca_zip_df
```

```python
#what zipcode have the most collision
most_common = ca_zip_df.zipcode.value_counts()
most_common
```

```python
#create summary bar data frame
most_common_df = pd.DataFrame({
    "count": most_common,
    
    })
most_common_df
```

```python
#sort top 7 zipcode
top_seven_zipcode = most_common_df.nlargest(7,'count')
top_seven_zipcode
```
